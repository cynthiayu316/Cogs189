```python
import numpy as np
import matplotlib.pyplot as plt

root = '/Users/katherinequan/Documents/GitHub/Cogs189/'
cur_file=root+str('output_after_sliding.txt')
#print(cur_file)

#load txt file data samples
with open(cur_file) as data:
    content = data.readlines()

element_list=[] # initialize a list


for i in range(len(content)):
    content[i] = content[i].strip('\n')
    content[i] = content[i].strip('[')
    content[i] = content[i].strip(']')
    content[i] = content[i].split(", ")
    content[i]=[float(c) for c in content[i]]

    #print(content[i])
    
dt_list_1 = content[0]
dt_list_2 = content[1]
dt_list_3 = content[2]
dt_list_4 = content[3]

#print(dt_list_4)
```

```python
# Calculate the average
highlighted_icon_mean = np.mean(dt_list_1)
inverted_object_mean = np.mean(dt_list_2)
inverted_eyelessface_mean = np.mean(dt_list_3)
inverted_face_mean = np.mean(dt_list_4)


# Calculate the standard deviation
highlighted_icon_std = np.std(dt_list_1)
inverted_object_std = np.std(dt_list_2)
inverted_eyelessface_std = np.std(dt_list_3)
inverted_face_std = np.std(dt_list_4)


# Define labels, positions, bar heights and error bar heights
labels = ['highlighted_icon', 'inverted_object', 'inverted_eyelessface', 'inverted_face']
x_pos = np.arange(len(labels))
# [highlighted_icon_mean, inverted_object_mean, inverted_eyelessface_mean, inverted_face_mean]
CTEs = [36.6074, 41.45, 42.31, 40.203]
print(CTEs)
error = [highlighted_icon_std, inverted_object_std, inverted_eyelessface_std, inverted_face_std]
```

```python
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs,
       yerr=error,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=30)
ax.set_ylabel('Distribution for 4 labels')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('Error Bar Chart for 4 types of objects')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
fig.set_size_inches(8.5, 5.5)
#plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
